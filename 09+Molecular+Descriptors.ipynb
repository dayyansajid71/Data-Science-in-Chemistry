```python
def calculate_descriptors(smiles):
    mol = Chem.MolFromSmiles(smiles)
    
    mw = Descriptors.MolWt(mol)
    logp = Descriptors.MolLogP(mol)
    hbd = Descriptors.NumHDonors(mol)
    hba = Descriptors.NumHAcceptors(mol)
    tpsa = Descriptors.TPSA(mol)
    SP3 = Descriptors.FractionCSP3(mol)
    
    print(f"Calculated Descriptors for {smiles}\nMolecular Weight: {mw}\nLogP: {logp}\nNo. of H Donors: {hbd}\nNo. of H Acceptors: {hba}\nTopological Surface Area: {tpsa}\nSP3 C %: {SP3}")
```

```python
calculate_descriptors(aspirin)
```

### From Descriptors to Rules

#### - Lipinski's Rule of Five (RO5)

- Molecular Weight (MW) ≤ 500
- LogP (octanol-water partition coefficient) ≤ 5
- Hydrogen Bond Donors (HBD) ≤ 5
- Hydrogen Bond Acceptors (HBA) ≤ 10

```python
def lipinski_RO5(smiles):
    mol = Chem.MolFromSmiles(smiles)
    
    mw = Descriptors.MolWt(mol)
    logp = Descriptors.MolLogP(mol)
    hbd = Descriptors.NumHDonors(mol)
    hba = Descriptors.NumHAcceptors(mol)
    
    satisfies_ro5 = (mw <= 500) and (logp <= 5) and (hbd <= 5) and (hba <= 10)
    
    return satisfies_ro5
```

```python
lipinski_RO5(aspirin)
```

```python
lipinski_RO5(insulin)
```

#### - Pfizer rule 3/75

- LogP (octanol-water partition coefficient) ≥ 3
- Topological polar surface area (TPSA) ≤ 75

```python
def pfizer_3_75(smiles):
    mol = Chem.MolFromSmiles(smiles)
    
    logp = Descriptors.MolLogP(mol)
    tpsa = Descriptors.TPSA(mol)
    
    if logp > 3 and tpsa < 75:
        result = "toxic"
    else:
        result = "not toxic"
        
    return result
```

```python
pfizer_3_75(aspirin)
```
