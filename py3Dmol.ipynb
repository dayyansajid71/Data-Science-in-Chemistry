```python
import pandas as pd
import py3Dmol as p3d
from rdkit import Chem
from rdkit.Chem import AllChem
import ipywidgets as w
from ipywidgets import interact
```

```python
@interact
def R3D(SMILES='C=O'):

    mol = Chem.MolFromSmiles(SMILES)
    display(mol)
    mol = Chem.AddHs(mol)
    AllChem. EmbedMolecule(mol)
    molblock = Chem.MolToMolBlock(mol)

    view = p3d.view(width=450, height=450)
    view.addModel(molblock, 'mol')
    view.setBackgroundColor("#323232")
    view.setStyle({"stick": {}, "sphere": {"scale": 0.3}})
    view.setViewStyle({"style": "outline", "color": "white", "width": 0.01})
    view.show()
```

```python
# Reading the dataset
df = pd.read_excel("new_silico_dataset.xlsx")

# Creating a mapping between molecule names and molecule smiles
molecule_map = {name: smiles for name, smiles in zip(df['Name'], df['SMILES'])}

# Creating dropdown widget with molecule names
dropdown = w.Dropdown(
    options=df['Name'].tolist(),
    description="Molecule:"
)

def R4D(SMILES, H=True, animate=False, link=True):

    Mol = Chem.MolFromSmiles(SMILES)
    if H == True:
        Mol = Chem.AddHs(Mol)
    elif H == False:
        pass
    AllChem.EmbedMolecule(Mol)
    AllChem.MMFFOptimizeMolecule(Mol)
    MOLBLOCK = Chem.MolToMolBlock(Mol)

    RENDER3D = p3d.view(width=880, height=220, viewergrid=(1, 4), linked=link)
    RENDER3D.addModel(MOLBLOCK)


    RENDER3D.setBackgroundColor("#745085", viewer=(0,0))
    RENDER3D.setStyle({'line': {'linewidth': 5}}, viewer=(0,0))

    RENDER3D.setBackgroundColor("#323232", viewer=(0,1))
    RENDER3D.setStyle({"stick": {'radius': 0}, "sphere": {"scale": 0}}, viewer=(0,1))
    RENDER3D.setViewStyle({"style": "outline", "color": "#CA86FF", "width": 0.05}, viewer=(0,1))

    RENDER3D.setBackgroundColor("#8f69fb", viewer=(0,2))
    RENDER3D.setStyle({"stick": {}, "sphere": {"scale": 0.3}}, viewer=(0,2))
    RENDER3D.setViewStyle({"style": "outline", "color": "#323232", "width": 0.05}, viewer=(0,2))

    RENDER3D.setBackgroundColor("#ccade1", viewer=(0,3))
    RENDER3D.setStyle({"sphere": {}}, viewer=(0,3))
    RENDER3D.setViewStyle({"style": "outline", "color": "#323232", "width": 0.05}, viewer=(0,3))

    if animate == True:
        RENDER3D.spin(True)
    elif animate == False:
        RENDER3D.spin(False)

    RENDER3D.zoomTo()

    return RENDER3D

def on_select(molecule_name,):
    molecule = molecule_map[molecule_name]
    return R4D(molecule)

interact(
    on_select,
    molecule_name=dropdown,
)
```
