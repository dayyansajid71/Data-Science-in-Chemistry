# Interactivity in Notebooks

```python
import ipywidgets as w
from ipywidgets import interact
from IPython.display import display
```

## ipywidgets Usecases

### - Download SDF from PubChem

```python
import pubchempy as pcp

def get_sdf(molecule_name):
    try:
        pcp.download("SDF", f"./{molecule_name}.sdf", molecule_name, "name", overwrite=True)
        print(f"Done. {molecule_name} SDF has been downloaded successfully.")
    
    except:
        print("Error!")
        
molecule_name_input = w.Text(value="", placeholder="Enter Molecule Name", description="Molecule:")

submit_button = w.Button(description="Get SDF", button_style="", tooltip="Click to download SDF")

def on_button_click(b):
    molecule_name = molecule_name_input.value
    get_sdf(molecule_name)

submit_button.on_click(on_button_click)

display(molecule_name_input, submit_button)
```

### - Visualizing 3D Structures from a database

```python
import pandas as pd
import imolecule
```

```python
df = pd.read_csv("pubchem_molecules.csv")
df
```

```python
dropdown = w.Dropdown(options=df["SMILES"].tolist(), description="Molecule")

def irender(molecule):
    return imolecule.draw(molecule, size=(400,400), drawing_type="wireframe", shader="toon", show_save=True)

interact(irender, molecule=dropdown)
```

```python
molecule_map = {name: smiles for name, smiles in zip(df["Name"], df["SMILES"])}

def on_select(molecule_name):
    molecule_smiles = molecule_map[molecule_name]
    print(molecule_smiles)
    
on_select("aspirin")
```

```python
dropdown = w.Dropdown(options=df["Name"].tolist(), description="Molecule")

def irender(molecule):
    return imolecule.draw(molecule, size=(400,400), drawing_type="wireframe", shader="toon", show_save=True)

molecule_map = {name: smiles for name, smiles in zip(df["Name"], df["SMILES"])}

def on_select(molecule_name):
    molecule_smiles = molecule_map[molecule_name]
    return irender(molecule_smiles)

interact(on_select, molecule_name=dropdown)
```

```python
dropdown = w.Dropdown(options=df["Name"].tolist(), description="Molecule")

def irender(molecule, drawing_type):
    return imolecule.draw(molecule, size=(400,400), drawing_type=drawing_type, shader="toon", show_save=True)

molecule_map = {name: smiles for name, smiles in zip(df["Name"], df["SMILES"])}

toggle_buttons = w.ToggleButtons(options=["wireframe", "ball and stick", "space filling"])

def on_select(molecule_name, drawing_type):
    molecule_smiles = molecule_map[molecule_name]
    return irender(molecule_smiles, drawing_type)

interact(on_select, molecule_name=dropdown, drawing_type=toggle_buttons)
```
