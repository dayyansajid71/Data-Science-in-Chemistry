### Search for a Compound

#### by Name

```python
for result in cs.search("caffeine"):
    print(result, result.smiles, result.common_name)
```

```python
name = cs.filter_name("ethanol")
```

```python
name
```

```python
result = cs.filter_results(name)
result
```

```python
for i in result:
    cmpd = cs.get_compound(i)
    print(cmpd.common_name)
```

#### by Formula

```python
formula = cs.filter_formula("C2H6O")
formula
```

```python
results = cs.filter_results(formula)
results
```

```python
for i in results:
    cmpd = cs.get_compound(i)
    print(cmpd.molecular_formula, cmpd.common_name)
```

```python
formula_batch = cs.filter_formula_batch_results(cs.filter_formula_batch(['C2H6O', 'C6H6O', 'C6H6N']))
formula_batch
```

#### by SMILES

```python
smiles = cs.filter_results(cs.filter_smiles("CCCCCC"))
smiles
```

```python
for i in smiles:
    cmpd = cs.get_compound(i)
    print(cmpd.molecular_formula, cmpd.common_name)
```

### Creating a database for a list of Molecules

```python
import pandas as pd

data = []

for name in molecules:
    compound = cs.search(name)[0]
    data.append({
        "Name" : name,
        "ChemSpider_ID" : compound.csid,
        "Common_Name" : compound.common_name,
        "Molecular_Formula" : compound.molecular_formula,
        "Molecular_Weight" : compound.molecular_weight,
        "SMILES" : compound.smiles,
        "InChI" : compound.inchi,
        "InChIKey" : compound.inchikey
    })
    
df = pd.DataFrame(data)
df
```

```python
df.to_csv("chempider_molecules.csv")
```

### Getting MOL/SDF 

```python
import os

newfolder = input("Create a New Folder for your Molecules:")
if not os.path.exists(newfolder):
    os.makedirs(newfolder, exist_ok=True)
    print(f"Folder created: {newfolder}")
else:
    print(f"Folder already exists: {newfolder}")
```

```python
sdfs = []

for name in molecules:
    compound = cs.search(name)[0]
    with open(f"{newfolder}\{name}.mol", "w") as f:
        f.write(compound.mol_3d)
    sdfs.append(compound.mol_3d)
    print(f"{compound.common_name} MOL file has been created successfully.")
    
with open("chemspider_molecules.sdf", "w") as f:
    f.write("".join(sdfs))
    
print("All Done!")    
```
