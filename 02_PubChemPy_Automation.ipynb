## 1. Automate Getting SDF of a list of Molecules

```python
pcp.download("SDF", "aspirin.sdf", "aspirin", "name")
```

```python
import os

newfolder = input("Create a New Folder for your Molecules:")
if not os.path.exists(newfolder):
    os.makedirs(newfolder, exist_ok=True)
    print(f"Folder created: {newfolder}")
else:
    print(f"Folder already exists: {newfolder}")
```

```python
for name in molecules:
    pcp.download("SDF", f"./{newfolder}/{name}.sdf", name, "name")
```

```python
pcp.get_sdf("aspirin", "name")
```

```python
def make_sdf(molecules):
    """
    molecules : a list of molecules
    """
    sdfs = []
    
    for name in molecules:
        cmpd_sdf = pcp.get_sdf(name, "name")
        sdfs.append(cmpd_sdf)
        
    with open("pubchem_molecules.sdf", "w") as f:
        f.write("".join(sdfs))
    print("DONE.")    

make_sdf(molecules)    
```

## 2. Creating a Database for a list of Molecules

```python
data = []
for name in molecules:
    compound = pcp.get_compounds(name, "name")[0]
    data.append({
        "Name": name,
        "CID": compound.cid,
        "IUPAC_Name": compound.iupac_name,
        "Molecular_Formula": compound.molecular_formula,
        "Molecular_Weight": compound.molecular_weight,
        "SMILES": compound.isomeric_smiles,
        "InChI": compound.inchi,
        "InChIKey": compound.inchikey
    })

data    
```

```python
import pandas as pd

df = pd.DataFrame(data)
df
```

```python
df.to_csv("pubchem_molecules.csv")
```
