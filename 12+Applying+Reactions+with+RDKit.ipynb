# Applying Reactions with RDKit

## - Syntax for Reaction SMARTS

### [Reactants]>>[Products]
____________________________________________

#### Example: Transforming a carbonyl [C=O] to an alcohol [C-OH]
- [C:2]=[O:1]>>[C:2][OH:1]
____________________________________________

### Limitations
- Complexity and Missing Specificity
- No-Matter Conditions

```python
from rdkit import Chem
from rdkit.Chem import AllChem, Draw
```

### - Amide Coupling Reaction

```python
amide_smarts = "[C:1](=[O:2])O.[N:3]>>[C:1](=[O:2])[N:3]"
```

```python
amide_rxn = AllChem.ReactionFromSmarts(amide_smarts)
amide_rxn
```

```python
benzoic = Chem.MolFromSmiles("C1=CC=C(C=C1)C(=O)O")
ethylamine = Chem.MolFromSmiles("CCN")
```

```python
Draw.MolsToGridImage([benzoic, ethylamine], molsPerRow = 2, subImgSize = (200, 200))
```

```python
products = amide_rxn.RunReactants([benzoic, ethylamine])
products
```

```python
products[0][0]
```

```python
Chem.MolToSmiles(products[0][0])
```

### - Peptide Synthesis

```python
peptide_smarts = "[NH2:1][CH:2]([C:3])C(=O)[OH].[NH2:4][CH:5]([C:6])C(=O)[OH] >> [NH:1][CH:2]([C:3])C(=O)[NH:4][CH:5]([C:6])C(=O)O"
```

```python
peptide_rxn = AllChem.ReactionFromSmarts(peptide_smarts)
peptide_rxn
```

```python
valine = Chem.MolFromSmiles("CC(C)[C@@H](C(=O)O)N")
serine = Chem.MolFromSmiles("C([C@@H](C(=O)O)N)O")
```

```python
Draw.MolsToGridImage([valine, serine], molsPerRow = 2, subImgSize = (200, 200))
```

```python
products = peptide_rxn.RunReactants([valine, serine])
products[0][0]
```
