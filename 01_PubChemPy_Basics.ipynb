### Searching from the compound database

###### get_compounds("identifier", "namespace")

```python
aspirin = pcp.get_compounds("aspirin", "name")
```

```python
type(aspirin)
```

```python
for i in aspirin:
    print(i.iupac_name, i.isomeric_smiles, i.molecular_formula)
```

```python
ethanol = pcp.get_compounds('CCO', 'smiles')
```

```python
for i in ethanol:
    print(f"Name: {i.iupac_name}\nSMILES: {i.isomeric_smiles}\nMolecular Formula: {i.molecular_formula}\nMolecular Weight: {i.molecular_weight}\nCID: {i.cid}")
```

```python
C6H6O = pcp.get_compounds('C6H6O', 'formula')
```

```python
C6H6O
```

```python
for i in C6H6O:
    print(f"Name: {i.iupac_name}\nSMILES: {i.isomeric_smiles}\nMolecular Formula: {i.molecular_formula}\nMolecular Weight: {i.molecular_weight}\nCID: {i.cid}\n__________________________________________")
```

### Getting Properties of a compound

The available properties are: 
"MolecularFormula", "MolecularWeight", "CanonicalSMILES", "IsomericSMILES", "InChI", "InChIKey", "IUPACName", "XLogP", "ExactMass", "MonoisotopicMass", "TPSA", "Complexity", "Charge", "HBondDonorCount", "HBondAcceptorCount", "RotatableBondCount", "HeavyAtomCount", "IsotopeAtomCount", "AtomStereoCount", "DefinedAtomStereoCount", "UndefinedAtomStereoCount", "CovalentUnitCount", "Volume3D", "ConformerCount3D"

```python
prop = ["MolecularFormula", "MolecularWeight", "CanonicalSMILES", "IsomericSMILES", "InChI", "InChIKey", "IUPACName", "XLogP", "ExactMass", "MonoisotopicMass", "TPSA", "Complexity", "Charge", "HBondDonorCount", "HBondAcceptorCount", "RotatableBondCount", "HeavyAtomCount", "IsotopeAtomCount", "AtomStereoCount", "DefinedAtomStereoCount", "UndefinedAtomStereoCount", "CovalentUnitCount", "Volume3D", "ConformerCount3D"]
```

```python
aspirin_ = pcp.get_properties(prop, "aspirin", "name")
```

```python
aspirin_
```
