```python
import ipywidgets as w
from ipywidgets import interact, interact_manual
```

## Interactive Function

```python
def square(x):
    return x**2

interact(square, x=5)
```

```python
def square(x=5):
    return x**2

interact(square)
```

```python
def formula(C,H,N):
    return(f"C{C}H{H}N{N}")

interact(formula, C=2, H=6, N=w.fixed(1))
```

```python
@interact_manual
def square(x=5):
    return x**2
```

```python
@interact(x=5)
def square(x):
    return x**2
```

```python
def hi(name, age):
    print(f"My name is {name} and I'm {age}")

interact(hi, name="Mohamed", age=28)    
```

| Keyword argument                                            | Widget      |
| ----------------------------------------------------------- | ----------- |
| True or False                                               | Checkbox    |
| 'Hi'                                                  | Text        |
| value or (min,max) or (min,max,step) if integers are passed | IntSlider   |
| value or (min,max) or (min,max,step) if floats are passed   | FloatSlider |
| ['A','B'] or [('one', 1), ('two', 2)]                  | Dropdown    |

```python
@interact
def f(x=['A','B']):
    return(x)
```

## Widgets

### - IntSlider

```python
slider = w.IntSlider(value=4, min=0, max=10, step=2, description="integer slider", orientation="horizontal")
slider
```

```python
current_value = slider.value
current_value
```

### - FloatSlider

```python
w.FloatSlider(value=7.5, min=0.5, max=10.5, step=0.1, description="float slider", orientation="horizontal")
```

### - IntRangeSlider

```python
int_range = w.IntRangeSlider(value=(20,40), min=0, max=100, step=2, description="Range Integer")
int_range
```

### - FloatRangeSlider

```python
w.FloatRangeSlider(value=(20.2,40.4), min=0, max=100, step=2.4, description="Range Float")
```

### - Text

```python
w.Text(value="Hello", placeholder="say hi", description="Text")
```

### - Button

### - ToggleButton

### - Interactive 3D Visualization of SMILES

```python
import imolecule

@interact
def irender(smiles="C=O"):
    return imolecule.draw(smiles, size=(400, 400), drawing_type="wireframe", shader="toon", show_save=True)
```

## Handling Button Click Events

```python
def on_button_click(b):
    print("Hi!")
    
button = w.Button(description="Hello!")

button.on_click(on_button_click)

button
```

```python
def on_button_click(b):
    def hi():
        print("Hi!")
    hi()
    
button = w.Button(description="Hello!")

button.on_click(on_button_click)

button
```

```python
name = w.Text(value="", description="Name")

def on_button_click(b):
    print(f"Hi! {name.value}")
    
button = w.Button(description="Hello!")

button.on_click(on_button_click)

display(name, button)
```
